##  문장 내의 단어들을 임베딩
아래 조건으로 영문 문장 4개 리스트를 Embedding 작성
- keras.layers.Embedding 사용
- 영문 문장은 100자 적용

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential

# Define the sentences
sentences = ["I love my dog", "I love my cat", "You love my dog!", "Do you think my dog is amazing?"]

# Tokenize the sentences
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>") # num_words is the size of the vocabulary
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(sentences)

# Pad the sequences
padded = pad_sequences(sequences, maxlen=100, padding='post') # maxlen is now set to 100

# Define the embedding layer
embedding = Embedding(10000, 16, input_length=100) # input_length is now set to 100

# Create the model and add the embedding layer
model = Sequential([embedding])

# Get the embeddings
embeddings = model.predict(padded)

print(embeddings)
```
