Pipeline = prompt + model + output parser

- role : system을 셋팅하는 곳 확인 필요

```python
topic = "고래를 주제로"
topic = "데이터분석가 주제로"
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

# prompt = ChatPromptTemplate.from_template("재미난 이야기 3줄 작성 {topic}")
prompt = ChatPromptTemplate.from_template("{topic} 과정 제목을 3개 작성")
model = ChatOpenAI()
output_parser = StrOutputParser()

# pipe line
chain = prompt | model | output_parser

chain.invoke({"topic": topic})
```

```python
type(prompt), type(model), type(output_parser)
```

Prompt

```python
prompt_value = prompt.invoke({"topic": topic})
prompt_value
```

```python
prompt_value.to_messages()
```

```python
prompt_value.to_string()
```

Model

```python
message = model.invoke(prompt_value)
message
```

```python
from langchain.llms import OpenAI

llm = OpenAI(model="gpt-3.5-turbo-instruct")
llm.invoke(prompt_value)
```

Output parser

```python
output_parser.invoke(message)
```

Entire Pipeline

```python
input = {"topic": topic}

prompt.invoke(input)
# > ChatPromptValue(messages=[HumanMessage(content='tell me a short joke about ice cream')])

(prompt | model).invoke(input)
# > AIMessage(content="Why did the ice cream go to therapy?\nBecause it had too many toppings and couldn't cone-trol itself!")
```

```python
from openai import OpenAI
client = OpenAI()

response = client.chat.completions.create(
  model="gpt-3.5-turbo-0301",
  messages=[
    {
      "role": "system",
      "content": "나는 편의점에 있다"
    },
    {
      "role": "user",
      "content": "우유에 대한 3요소 설명"
    }
  ],
  temperature=1,      # 예측 불가한 조합 구성
  max_tokens=256,
  top_p=1,            # 추천 확률 임계치 설정
  frequency_penalty=0,
  presence_penalty=0
)
```
